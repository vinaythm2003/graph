```python
%matplotlib notebook
import pandas as pd
from matplotlib import pyplot as plt
import pandas.io.sql
import pyodbc
# Parameters
server = 'DESKTOP-50B13P8\VINAY'
db = 'DemoDB'

# Create the connection
conn = pyodbc.connect('DRIVER={SQL Server};SERVER=' + server + ';DATABASE=' + db + ';Trusted_Connection=yes')

# query db
sql = """

Select TimeLogged, Status from DemoDB.dbo.Demo1 where AppDate = '2019-10-05' and AppName like '%DSD%'

"""
df = pandas.io.sql.read_sql(sql, conn)
df.head()

sql1 = """

Select TimeLogged, Status from DemoDB.dbo.Demo1 where AppDate = '2019-10-05' and AppName like '%NLP%'

"""
df1 = pandas.io.sql.read_sql(sql1, conn)
df1.head()

sql2 = """

Select TimeLogged, Status from DemoDB.dbo.Demo1 where AppDate = '2019-10-05' and AppName like '%XIW%'

"""
df2 = pandas.io.sql.read_sql(sql2, conn)
df2.head()

plt.plot(df.TimeLogged, df.Status)
plt.plot(df1.TimeLogged, df1.Status)
plt.plot(df2.TimeLogged, df2.Status)
plt.legend(['DSD Tab', 'New Landing Page', 'Workshop'])
plt.xlabel('Hours')
plt.ylabel('Status')
```
